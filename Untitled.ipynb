```python
import ctypes
import time

# Constantes para simular tecla F15 (0x7E) - tecla que não faz nada visível
VK_F15 = 0x7E

# Estruturas necessárias para SendInput
PUL = ctypes.POINTER(ctypes.wintypes.ULONG)

class KEYBDINPUT(ctypes.Structure):
    _fields_ = [("wVk", ctypes.wintypes.WORD),
                ("wScan", ctypes.wintypes.WORD),
                ("dwFlags", ctypes.wintypes.DWORD),
                ("time", ctypes.wintypes.DWORD),
                ("dwExtraInfo", PUL)]

class HARDWAREINPUT(ctypes.Structure):
    _fields_ = [("uMsg", ctypes.wintypes.DWORD),
                ("wParamL", ctypes.wintypes.WORD),
                ("wParamH", ctypes.wintypes.WORD)]

class MOUSEINPUT(ctypes.Structure):
    _fields_ = [("dx", ctypes.wintypes.LONG),
                ("dy", ctypes.wintypes.LONG),
                ("mouseData", ctypes.wintypes.DWORD),
                ("dwFlags", ctypes.wintypes.DWORD),
                ("time", ctypes.wintypes.DWORD),
                ("dwExtraInfo", PUL)]

class INPUT(ctypes.Union):
    _fields_ = [("ki", KEYBDINPUT),
                ("mi", MOUSEINPUT),
                ("hi", HARDWAREINPUT)]

class INPUT_STRUCT(ctypes.Structure):
    _fields_ = [("type", ctypes.wintypes.DWORD),
                ("ii", INPUT)]

def press_key(key):
    extra = ctypes.pointer(ctypes.wintypes.ULONG(0))
    ii_ = INPUT(type=1, ki=KEYBDINPUT(key, 0, 0, 0, extra))
    input_struct = INPUT_STRUCT(type=1, ii=ii_)
    ctypes.windll.user32.SendInput(1, ctypes.pointer(input_struct), ctypes.sizeof(input_struct))

def release_key(key):
    extra = ctypes.pointer(ctypes.wintypes.ULONG(0))
    ii_ = INPUT(type=1, ki=KEYBDINPUT(key, 0, 0x0002, 0, extra))  # KEYEVENTF_KEYUP
    input_struct = INPUT_STRUCT(type=1, ii=ii_)
    ctypes.windll.user32.SendInput(1, ctypes.pointer(input_struct), ctypes.sizeof(input_struct))

def simulate_f15():
    press_key(VK_F15)
    release_key(VK_F15)

try:
    while True:
        simulate_f15()
        time.sleep(60)  # Ajuste para menos (ex: 30) se o Teams for muito sensível
except KeyboardInterrupt:
    print("\n🛑 Simulação parada. Status voltará ao normal.")
```
